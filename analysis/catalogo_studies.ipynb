```python
import pandas as pd
```

```python
lots = pd.read_csv('../clean-files/catalogo_artworks_info.csv')
lots.columns
```

```python
lots.dropna(subset=['Price (BRL)', 'Medium'], inplace=True)

guache = lots[lots['Medium'].str.contains('guache', case=False)]
cartao = lots[lots['Medium'].str.contains('cartão', case=False)]
papel = lots[lots['Medium'].str.contains('papel', case=False)]
tela = lots[lots['Medium'].str.contains('tela', case=False)]
oleo = lots[lots['Medium'].str.contains('óleo', case=False)]
acr = lots[lots['Medium'].str.contains('acrílico', case=False)]
serigrafia = lots[lots['Medium'].str.contains('serigrafia', case=False)]
desenho = lots[lots['Medium_type'] == 'desenho']

for medium in [guache, cartao, papel, tela, oleo, acr, serigrafia, desenho]:
    print('mean price :' + str(medium['Price (BRL)'].mean()))
    print('std price :' + str(medium['Price (BRL)'].std()))
    print('min price :' + str(medium['Price (BRL)'].min()))
    print('max price :' + str(medium['Price (BRL)'].max()))
    print('---------')
```
